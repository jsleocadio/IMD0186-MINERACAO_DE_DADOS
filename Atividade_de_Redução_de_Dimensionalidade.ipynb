```python
import pandas as pd
```

```python
BaseOriginal = pd.read_csv("Image.csv")
labels = BaseOriginal['Class']
BaseOriginal = BaseOriginal.drop(columns=['Class'])
BaseOriginal.head()
```

## 2. Avalie a correlação entre os atributos e crie duas bases reduzidas,

```python
def get_correlation_df(df, nsmallest):
  import numpy as np
  # Get the correlation matrix
  correlation_matrix = df.corr()
  # Set diagonal values to NaN to exclude self-correlations
  np.fill_diagonal(correlation_matrix.values, np.nan)
  # Unstack the correlation matrix and sort by correlation values
  sorted_correlations = correlation_matrix.unstack().sort_values(ascending=False)
  # Filter for correlations greater than 0
  positive_correlations = sorted_correlations[sorted_correlations != 0]
  # Get the n smallest positive correlations
  smallest_positive_correlations = positive_correlations.nsmallest(nsmallest)
  # Get the columns corresponding to the smallest positive correlations
  columns_base = []
  for i in range(len(smallest_positive_correlations.index)):
    columns_base.append(smallest_positive_correlations.index[i][0])
  return df[columns_base]
```

### a. Uma base com os 3 atributos menos correlacionados (BaseReduzida-Corr-1); e

```python
BaseReduzidaCorr1 = get_correlation_df(BaseOriginal, 3)
BaseReduzidaCorr1.head()
```

### b. Uma base com os 9 atributos menos correlacionados (BaseReduzida-Corr-2).

```python
BaseReduzidaCorr2 = get_correlation_df(BaseOriginal, 9)
BaseReduzidaCorr2.head()
```

## 3. Aplique o PCA na base original


```python
def get_pca_df(df, n_components):
  from sklearn.decomposition import PCA
  pca = PCA(n_components=n_components)
  return pd.DataFrame(pca.fit_transform(df))
```

### a. Escolhendo 90% de cobertura mínima (Base-Reduzida-PCA-1); e

```python
BaseReduzidaPCA1 = get_pca_df(BaseOriginal, 0.9)
BaseReduzidaPCA1.head()
```

### b. Escolhendo as 3 componentes principais (Base-Reduzida-PCA-2).

```python
BaseReduzidaPCA2 = get_pca_df(BaseOriginal, 3)
BaseReduzidaPCA2.head()
```

# Para as cinco bases (Base-Original, Base-Reduzida-Corr-1, Base-Reduzida-Corr-2, Base-Reduzida-PCA-1 e Base-Reduzida-PCA-2), aplique o método k-NN (visto na aula de Aprendizado de Máquina em todas as bases. Analise a acurácia resultantes das cinco bases e responda as perguntas:


```python
def get_accuracy(df, labels):
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import train_test_split
  from sklearn.metrics import accuracy_score
  # Split the data
  x_train, x_test, y_train, y_test = train_test_split(df, labels, test_size=0.2, random_state=42)
  # Initialize KNN classifiers
  knn = KNeighborsClassifier()
  # Fit the models
  knn.fit(x_train, y_train)
  # Make predictions
  y_pred = knn.predict(x_test)
  # Calculate accuracy scores
  accuracy = accuracy_score(y_test, y_pred)
  return accuracy
```

```python
BaseOriginalAccuracy = get_accuracy(BaseOriginal, labels)
BaseReduzidaCorr1Accuracy = get_accuracy(BaseReduzidaCorr1, labels)
BaseReduzidaCorr2Accuracy = get_accuracy(BaseReduzidaCorr2, labels)
BaseReduzidaPCA1Accuracy = get_accuracy(BaseReduzidaPCA1, labels)
BaseReduzidaPCA2Accuracy = get_accuracy(BaseReduzidaPCA2, labels)
print(f"Base Original Accuracy: {BaseOriginalAccuracy}")
print(f"Base Reduzida Corr 1 Accuracy: {BaseReduzidaCorr1Accuracy}")
print(f"Base Reduzida Corr 2 Accuracy: {BaseReduzidaCorr2Accuracy}")
print(f"Base Reduzida PCA 1 Accuracy: {BaseReduzidaPCA1Accuracy}")
print(f"Base Reduzida PCA 2 Accuracy: {BaseReduzidaPCA2Accuracy}")
```

## a. O que aconteceu com a acurácia quando reduzimos a dimensionalidade usando a correlação?

A escolha de três features com menor correlação diferente de zero parece reduzir significativamente a acurácia. Essas features podem não conter informações suficientes ou relevantes para a tarefa de classificação, levando a um desempenho inferior.

Selecionar mais features (nove) com menor correlação ainda mostra uma redução na acurácia, mas é menos severa que na primeira redução. Isso indica que incluir mais features (mesmo que de baixa correlação) pode ajudar a capturar mais informações relevantes, melhorando a acurácia em relação à Base Reduzida Corr 1.

## b. O que aconteceu com a acurácia quando reduzimos a dimensionalidade usando o PCA?


Utilizar PCA para reter 90% da variância explica a maioria da variância nos dados, resultando em uma boa acurácia. Isso mostra que a redução dimensional com PCA pode ser eficaz, preservando a maioria das informações relevantes.

Reduzir para apenas três componentes principais com PCA resulta em uma acurácia significativamente mais baixa. Isso indica que três componentes não são suficientes para capturar as informações necessárias para a classificação, levando a um modelo menos eficaz.
